```python
import pandas as pd
import numpy as np

# ------------------------------------------------------------
# 1. Define boundary values based on your original data
#    (min and max for each column)
# ------------------------------------------------------------
boundaries = {
    'Internal_Avg_Sem1': (19.5000, 47.44),
    'Internal_Avg_Sem2': (16.0000, 48.50),
    'Internal_Avg_Sem3': (9.9545, 49.18),
    'Internal_Avg_Sem4': (12.3333, 48.71),
    'Internal_Avg_Sem5': (18.5909, 48.50),
    'SGPA_Sem1': (3.0435, 9.6957),
    'SGPA_Sem2': (2.6400, 10.0000),
    'SGPA_Sem3': (0.5000, 9.8182),
    'SGPA_Sem4': (0.4400, 10.0000),
    'SGPA_Sem5': (1.9546, 9.5909),
    'CGPA': (2.0118, 9.6715),
    'Ext_Avg_Sem1': (36.0000, 92.71),
    'Ext_Avg_Sem2': (30.5600, 96.12),
    'Ext_Avg_Sem3': (13.8182, 100),
    'Ext_Avg_Sem4': (5.3333, 93.57),
    'Ext_Avg_Sem5': (22.8182, 89.82),
    'Absent Percentage (%)': (0.0000, 37.42),
}

# Binary column probability (failure before sem5)
FAILURE_PROB = 0.30

# ------------------------------------------------------------
# 2. Helper: generate a value uniformly within a range
# ------------------------------------------------------------
def random_in_range(min_val, max_val, decimals=2):
    return round(np.random.uniform(min_val, max_val), decimals)

# ------------------------------------------------------------
# 3. Generate synthetic dataset (RANDOM each run)
# ------------------------------------------------------------
# 🔁 No seed set → different results every time
# (If you want reproducibility later, set a specific seed here)
# np.random.seed(42)   # <-- REMOVED or COMMENTED

n_records = 300

data = []

for _ in range(n_records):
    record = {}

    # Generate internal averages
    for sem in range(1, 6):
        col = f'Internal_Avg_Sem{sem}'
        record[col] = random_in_range(*boundaries[col], decimals=2)

    # Generate external averages
    for sem in range(1, 6):
        col = f'Ext_Avg_Sem{sem}'
        record[col] = random_in_range(*boundaries[col], decimals=2)

    # Generate SGPA values (independently)
    sgp_list = []
    for sem in range(1, 6):
        col = f'SGPA_Sem{sem}'
        val = random_in_range(*boundaries[col], decimals=4)
        record[col] = val
        sgp_list.append(val)

    # Compute CGPA as average of the 5 SGPAs
    record['CGPA'] = round(np.mean(sgp_list), 4)

    # Generate absent percentage
    record['Absent Percentage (%)'] = random_in_range(*boundaries['Absent Percentage (%)'], decimals=2)

    # Binary failure flag
    record['Has_Failure_Before_Sem5'] = np.random.binomial(1, FAILURE_PROB)

    data.append(record)

# Create DataFrame
df_synth = pd.DataFrame(data)

# ------------------------------------------------------------
# 4. Reorder columns to match original file
# ------------------------------------------------------------
original_columns = [
    'Has_Failure_Before_Sem5',
    'Internal_Avg_Sem1', 'Internal_Avg_Sem2', 'Internal_Avg_Sem3', 'Internal_Avg_Sem4', 'Internal_Avg_Sem5',
    'SGPA_Sem1', 'SGPA_Sem2', 'SGPA_Sem3', 'SGPA_Sem4', 'SGPA_Sem5',
    'CGPA',
    'Ext_Avg_Sem1', 'Ext_Avg_Sem2', 'Ext_Avg_Sem3', 'Ext_Avg_Sem4', 'Ext_Avg_Sem5',
    'Absent Percentage (%)'
]
df_synth = df_synth[original_columns]

# ------------------------------------------------------------
# 5. Save to CSV (use a timestamp or just overwrite)
# ------------------------------------------------------------
output_file = "combined_finals_with_attendance.csv"
df_synth.to_csv(output_file, index=False)
print(f"✅ Generated {len(df_synth)} synthetic records using boundary values.")
print(f"File saved as '{output_file}'")
```
